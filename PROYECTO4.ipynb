```python
# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.
messages["message_date"] = pd.to_datetime(messages["message_date"])

# 2. Crear una nueva columna para extraer el mes del mensaje
messages["month"] = messages["message_date"].dt.to_period("M")

# 3. Calcular el número de mensajes enviados por cada usuario al mes
monthly_messages_per_user = (
    messages
    .groupby(["user_id", "month"])["id"]
    .count()
    .reset_index()
    .rename(columns={"id": "n_messages"})
)

# 4. Mostrar un histograma del número de mensajes enviados por mes
monthly_messages_per_user
```
